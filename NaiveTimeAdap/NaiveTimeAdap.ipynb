```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 100
n = 50000000
Z1 = np.random.normal(0, np.sqrt(dt), n)
Z2 = np.random.normal(0, np.sqrt(dt / 2), n)
Z2 = np.random.normal(0, np.sqrt(dt / 2), n) + Z2
```

```python
fig = plt.figure()
plt.hist(Z1, bins=100)
plt.hist(Z2, bins=100, alpha = 0.5)
plt.show()
```
